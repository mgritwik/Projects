```python
import pandas as pd
import numpy as np
```

```python
cols=[ 'score',
 'text',
 'response_date_ist',
 'company_id',
 'company_name',
 'plan_name',
 'tier',
 'account_manager',
 'shipments_m3',
 'shipments_m2',
 'shipments_m1',
 'shipments_m0']
df_nov=pd.read_excel("NPS(June-till date).xlsx",sheet_name="Nov-24",usecols=cols)
# df_oct=pd.read_excel("NPS(June-till date).xlsx",sheet_name="Oct'20",usecols=cols)
# df_sep=pd.read_excel("NPS(June-till date).xlsx",sheet_name="Sep'20",usecols=cols)
# df_aug=pd.read_excel("NPS(June-till date).xlsx",sheet_name="Aug'20",usecols=cols)
# df_jul=pd.read_excel("NPS(June-till date).xlsx",sheet_name="July'20",usecols=cols)
# df_jun=pd.read_excel("NPS(June-till date).xlsx",sheet_name="June'20",usecols=cols)
```

```python
df_nov['month']='November'
# df_oct['month']='October'
# df_sep['month']='September'
# df_aug['month']='August'
# df_jul['month']='July'
# df_jun['month']='June'
# df_all = pd.concat([df_nov,df_oct,df_sep,df_aug,df_jul,df_jun], axis = 0)
df_all=df_nov
df_pivot = df_all.groupby(['month','company_id']).agg({'score':'mean',
                                                        'shipments_m0':'mean',
                                                      'shipments_m1':'mean',
                                                      'shipments_m2':'mean',
                                                      'shipments_m3':'mean'})
df_pivot=df_pivot.reset_index()
df_pivot['company_name']=""
df_pivot['plan_name']=""
df_pivot['tier']=""
df_pivot['account_manager']=""
for i in df_pivot.index:
    temp=df_all.loc[df_all['company_id'] == df_pivot.at[i,'company_id']]
    df_pivot.at[i,'company_name']=temp['company_name'].iloc[0]
    df_pivot.at[i,'plan_name']=temp['plan_name'].iloc[0]
    df_pivot.at[i,'tier']=temp['tier'].iloc[0]
    df_pivot.at[i,'account_manager']=temp['account_manager'].iloc[0]
# df_pivot.to_excel('NPS_All_months_data.xlsx', index=False)
```

```python
df_new = df_all[df_all.groupby('company_id').response_date_ist.transform('max') == df_all['response_date_ist']]
```

```python
df_new=df_new[['company_id','score']]
df_new.columns=['company_id', 'latest_score']
```
